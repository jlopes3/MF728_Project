```python
import pandas as pd
import numpy as np
from scipy.stats import pearsonr, spearmanr
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv('../merton_model_output.csv')
df_valid = df[df['merton_pd'].notna()].copy()
df_valid['log_E'] = np.log(df_valid['market_cap'])
```

### Pearson and Spearman correlations between market_cap and merton_pd

```python
pearson_corr, pearson_pval = pearsonr(df_valid['market_cap'], df_valid['merton_pd'])
spearman_corr, spearman_pval = spearmanr(df_valid['market_cap'], df_valid['merton_pd'])
print(f"Pearson r = {pearson_corr:.4f}, p = {pearson_pval:.4g}")
print(f"Spearman ρ = {spearman_corr:.4f}, p = {spearman_pval:.4g}")
```

### Pearson and Spearman correlations between log_E (log(market_cap)) and merton_pd

```python
pearson_corr, pearson_pval = pearsonr(df_valid['log_E'], df_valid['merton_pd'])
spearman_corr, spearman_pval = spearmanr(df_valid['log_E'], df_valid['merton_pd'])
print(f"Pearson r = {pearson_corr:.4f}, p = {pearson_pval:.4g}")
print(f"Spearman ρ = {spearman_corr:.4f}, p = {spearman_pval:.4g}")
```

```python
df = pd.read_csv('../merton_model_output.csv', parse_dates=['date'])
df.columns = df.columns.str.strip()

df = df[df['merton_pd'].notna()].copy()
df = df[['date', 'tic', 'market_cap', 'merton_pd']]
df = df.sort_values(['date', 'tic'])
df = df.groupby(['date', 'tic']).last().reset_index()

df_melt = df.melt(id_vars=['date', 'tic'], 
                  value_vars=['market_cap', 'merton_pd'], 
                  var_name='variable', 
                  value_name='value')

df_pivot = df_melt.pivot(index='date', columns=['tic', 'variable'], values='value')

df_pivot.columns.names = ['company', 'variable']

model_output = df_pivot.xs('merton_pd', axis=1, level='variable')

companies_with_nan = model_output.columns[model_output.isna().any()]
df_pivot = df_pivot.drop(columns=companies_with_nan, level=0)
```

```python

market_cap = df_pivot.xs('market_cap', axis=1, level='variable')
log_mkt_cap = np.log(market_cap)
log_returns = log_mkt_cap.diff()
for company in log_returns.columns:
    df_pivot[(company, 'log_return')] = log_returns[company]

neg_log_return = log_returns.clip(upper=0)
pos_log_return = log_returns.clip(lower=0)

for company in log_returns.columns:
    df_pivot[(company, 'neg_log_return')] = neg_log_return[company]
    df_pivot[(company, 'pos_log_return')] = pos_log_return[company]

df_pivot = df_pivot.sort_index(axis=1)

window_size = 252

log_returns_252 = log_returns.rolling(window=window_size, min_periods=1).sum()
for company in log_returns_252.columns:
    df_pivot[(company, 'log_returns_252')] = log_returns_252[company]

df_pivot = df_pivot.sort_index(axis=1)
```

```python
import statsmodels.api as sm

companies = df_pivot.columns.get_level_values(1).unique()

results = {}
companies = df_pivot.columns.get_level_values('company').unique()

for company in companies:
    try:
        y = df_pivot[(company, 'merton_pd')]

        X = df_pivot[[ (company, 'log_return'), (company, 'log_returns_252') ]]

        df_reg = pd.concat([y, X], axis=1).dropna()

        y_clean = df_reg[(company, 'merton_pd')]
        X_clean = df_reg[[ (company, 'log_return'), (company, 'log_returns_252')]]

        X_clean = sm.add_constant(X_clean)

        model = sm.OLS(y_clean, X_clean).fit()

        results[company] = model

    except Exception as e:
        print(f"Error with {company}: {e}")
```

```python
import matplotlib.pyplot as plt

betas_log_return = [results[c].params[(c, 'log_return')] for c in results.keys()]

plt.hist(betas_log_return, bins=30, edgecolor='black')
plt.title('Distribution of Log-Return Betas')
plt.xlabel('Beta')
plt.ylabel('Frequency')
plt.show()
```

```python
betas_cum_log_return = [results[c].params[(c, 'log_returns_252')] for c in results.keys()]
plt.hist(betas_cum_log_return, bins=30, edgecolor='black')
plt.title('Distribution of Cumulative Log-Return Betas')
plt.xlabel('Beta')
plt.ylabel('Frequency')
plt.show()
```

```python
results = {}

# Step 1: Create delta default probabilities
delta_merton_pd = df_pivot.xs('merton_pd', axis=1, level='variable').diff()

# Step 2: Create negative returns
log_returns = df_pivot.xs('log_return', axis=1, level='variable')
neg_log_return = log_returns.clip(upper=0)

# Step 3: Only keep companies that exist in both datasets
companies = delta_merton_pd.columns.intersection(neg_log_return.columns)

for company in companies:
    try:
        # Step 4: Extract dependent variable
        y = delta_merton_pd[company]

        # Step 5: Extract independent variable
        X = neg_log_return[[company]].rename(columns={company: 'neg_log_return'})
        X = sm.add_constant(X)

        # Step 6: Drop missing rows
        df_reg = pd.concat([y, X], axis=1).dropna()

        y_clean = df_reg[company]
        X_clean = df_reg[['const', 'neg_log_return']]

        # Step 7: Run OLS
        model = sm.OLS(y_clean, X_clean).fit()

        results[company] = model

    except Exception as e:
        print(f"Error with {company}: {e}")
```

```python
summary = []

for company, model in results.items():
    try:
        coef = model.params['neg_log_return']
        tstat = model.tvalues['neg_log_return']
        pval = model.pvalues['neg_log_return']

        summary.append({
            'company': company,
            'beta_neg_log_return': coef,
            't_stat': tstat,
            'p_value': pval
        })
        
    except Exception as e:
        print(f"Error processing {company}: {e}")

summary_df = pd.DataFrame(summary)


plt.figure(figsize=(10,6))
plt.hist(summary_df['beta_neg_log_return'], bins=30, edgecolor='black')
plt.axvline(0, color='red', linestyle='--')  # vertical line at 0
plt.title('Distribution of Betas on Negative Log-Returns')
plt.xlabel('Beta')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
significant_positive = summary_df[(summary_df['beta_neg_log_return'] > 0) & (summary_df['p_value'] < 0.05)]

print(f"{len(significant_positive)} companies out of {len(summary_df)} have a significantly positive beta (p < 0.05)")
```
